```python
# setup and your first plot
# matplot basics -> the foundation of all Python charts

# matplotlib is the core plotting library. Seaborn is built on top of it and makes beautifil statistical charts with much less code.
# the standard pattern for every plot:
# 1. Create the data
# 2. Call a plot function
# 3. Add labels and title
# 4. Show or save

# pli.tight_layout() prevents labels from being cut off. Always call it before plt.show() or plt.savefig()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Your first line chart

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 145, 132, 178, 195, 210]

plt.figure(figsize=(8,4))
plt.plot(months, sales, marker = "o", color="steelblue", linewidth=2)
plt.title("Monthly sales 2026")
plt.xlabel("Month")
plt.ylabel("Sales (units)")
plt.tight_layout()
plt.show()
```

```python
# creating a line chart of your own choice

week = ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]
daily_steps = [8000, 8500, 8200, 14000, 9000, 8500, 200]

plt.figure(figsize=(8,4))
plt.plot(week, daily_steps, marker="o", color="green", linewidth=4)
plt.title("Daily steps for a week")
plt.xlabel("Weeks")
plt.ylabel("Steps Count")
plt.tight_layout()
plt.show()
```

```python
# temperature across 6 cities

cities = ["Kathmandu", "Bhaktapur", "Lalitpur", "Pokhara", "Chitwan", "Butwal"]
temperature = [24, 20, 24, 28, 30, 30 ]

plt.figure(figsize=(8,4))
plt.plot(cities, temperature, marker="o", color="red", linewidth =3)
plt.title("Temperature across cities")
plt.xlabel("Cities")
plt.ylabel("Temperature ('C)")
plt.tight_layout()
plt.show()
```

```python
# Scores over 5 tests

test = ["test1", "test2", "test3", "test4", "test5"]
score = [20, 18, 12, 18, 19]

plt.figure(figsize=(8,4))
plt.plot(test, score, marker="o", color="orange", linewidth=2)
plt.title("Scores over 5 tests")
plt.xlabel("Tests")
plt.ylabel("Score")
plt.tight_layout()
plt.show()
```

```python
# bar charts : comparing categories visually
# bar charts are the most used chart type in data analytics. Use them to commpare values across categories

# plt.bar() -> vertical bars
# ply.barh() -> horizontal bars (bettter when category names are long)

# you can use sns.barplot() which automatically calculates means and add confidence intervals, great for grouped data/

df = pd.DataFrame({
    "subject": ["Maths", "Science", "English", "History", "Nepali"],
    "avg_score": [82, 76, 89, 71, 85]
})

# Matplotlib bar chart
plt.figure(figsize=(8,4))
plt.bar(df["subject"], df["avg_score"], color="steelblue")
plt.title("Average score by subject")
plt.xlabel("Subject")
plt.ylabel("Average score")
plt.tight_layout()
plt.show()

# Seaborn baro=plot - cleaner syntax with DataFrames
plt.figure(figsize=(8,4))
sns.barplot(data =df, x="subject", y="avg_score", palette="Blues_d")
plt.title("Average score by subject (Seaborn)")
plt.tight_layout()
plt.show()
```

```python
# Histograms and scatter plots: distributions and relationships
# two chart types that are essential for understanding data:

# Histogram -> show the distribution if single numeric variable. How many values fall in each range? Are they normally distributed? Skewed?

# Scatter plot -> shows the relationship between two numeric variables. Is there a correlation? Any clusters or outliers?

# These are the two charts you'll use most during exploartory data analysis (EDA)
# in seaborn, sns.histplot(kde=True) adds a smooth density curve on top of histogram, much more informative than a plain histogram
```

```python
scores = np.random.normal(loc=75, scale =12, size =200)

# Histograms
plt.figure(figsize=(8,4))
sns.histplot(scores, bins=20, kde=True, color="steelblue")
plt.title("Distribution of student scores")
plt.xlabel("Score")
plt.tight_layout()
plt.show()

# Scatterplot

study_hours =np.random.uniform(1,10,50)
exam_score = study_hours*8 + np.random.normal(0,5,50)

plt.figure(figsize=(8,5))
plt.scatter(study_hours, exam_score, color="coral", alpha=0.7)
plt.title("Study hours vs exam score")
plt.xlabel("Study hours")
plt.ylabel("Exam score")
plt.tight_layout()
plt.show()
```

```python
# generating 100 random heights, plot a histogram with kde =True and scatterplot if height vs weight

ran_height = np.random.normal(loc=165, scale=8, size=100)
weight = ran_height * 0.4 + np.random.normal (0,5,100)

# histogram
plt.figure(figsize=(8,4))
sns.histplot(ran_height, bins =20, kde =True, color="red")
plt.title("Plot of random heights")
plt.xlabel("heights")
plt.tight_layout()
plt.show()

# scatter plot
plt.figure(figsize=(8,4))
plt.scatter(ran_height, weight, color="blue", alpha=0.7)
plt.title("Random height vs weight")
plt.xlabel("Random height")
plt.ylabel("Weight")
plt.show()
```

```python
# seaborn statistical charts: beautiful charts with less code
# seborn shines when working with dataframes. three charts that come up constantly in data analytics:
# 1. sns.boxplot() -> shows median, quartiles, and outliers for each category
# 2. sns.heatmap() -> shows correlations between all numeric columns as a colour grid
# 3. sns.pairlot() -> plots ecery numeric column against every other, grat for quick EDA
# heatmap of correlations is one of the first thing data analysts run on new dataset.

# df.corr() computes the correlation matrix. Pass it straight into sns.heatmap(annot=True) to see all relationship at once.
```

```python
df = pd.DataFrame({
    "score": [88, 72, 95, 61, 79, 85, 68, 91],
    "hours": [5, 3, 7, 2, 4, 6,3, 7],
    "absences": [1, 2, 0, 6, 2, 1, 5, 0],
    "subject": ["Maths", "Science", "Maths", "Science",
                 "Maths", "Science", "Maths", "Science"],
    })

# boxplot -> score distribution per subject
plt.figure(figsize=(7,4))
sns.boxplot(data= df, x="subject", y="score", palette="Set2")
plt.title("Score distribution by subject")
plt.tight_layout()
plt.show()

# heatmap -> Correlation between numeric columns
plt.figure(figsize=(6,4))
sns.heatmap(df[["score","hours","absences"]].corr(),
            annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation heatmap")
plt.tight_layout()
plt.show()
```

```python
# subplots : multiple charts together
# putting charts side by side
# In real data naalytics report you often want multiple charts on one figure. plt.subplot() creates a grid of axes -> you specify rows and cols, then plot on each axis individually
#each subplot tells part of the story.

# use fig.axes = plt.subplot(1,2, figsize=(!2,4)) for side by side charts. Access each chart with axes[0] and axes[1]
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,4))

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 145, 132, 178, 195, 210]
costs = [80, 90, 85, 100, 110, 120]

# Left chart - line chart
axes[0].plot(months, sales, marker="o", color="steelblue", label="Sales")
axes[0].plot(months, costs, marker="s", color="coral", label="Costs")
axes[0].set_title("Sales vs costs")
axes[0].set_xlabel("Month")
axes[0].legend()

# Right chart - bar chart
profit = [s-c for s,c in zip(sales,costs)]
axes[1].bar(months, profit, color="green")
axes[1].set_title("Monthly profit")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Profit")
plt.tight_layout()
plt.show()
```
