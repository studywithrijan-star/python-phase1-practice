```python
# group by: summarising by category

# groupby splits data into groups and lers you calulte statistics for each group. Think of it like a pivot table in Excel
# the pattern is always df.groupby("column")["value_col"].function()

# agg() lets you calculate multiple statistics at once -mean, min, max, count, all in one line
```

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Aarav", "Priya", "Sita", "Ram", "Rijan", "Gita"],
    "subject": ["Maths", "Science", "Maths", "Science", "Maths", "Science"],
    "score": [88, 73, 95, 79, 85, 61],
    "city": ["Kathmandu", "Pokhara", "Kathmandu", "Pokhara", "Kathmandu", "Biratnagar"]
})

# Average score per subject
print(df.groupby("subject")["score"].mean().round(1))

# multiple stats ar once
print(df.groupby("subject")["score"].agg(["mean", "min", "max", "count"]).round(1))
```

```python
df = pd.read_csv("learn.csv")

print(f"The csv file is as:\n{df}\n")

# checking nulls and nan values
print(df.isnull().sum())

# Handling missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
df["Passed"] = df["Passed"].fillna("Unknown")

print(f"The file after handling missing values:\n{df}\n")

# average per score per passed
print(df.groupby("Passed")["Marks"].mean().round(1))

print(df.groupby("Passed")["Marks"].max())
```

```python
# groupby on multiple column
# can group by more than one column ata once by passing a list.
# very useful for comparing categories
# when grouping multiple column, the results has a multi-levle index.
# use .reset.index() at the end to flatten it back to a regular DataFrame
```

```python
df = pd.DataFrame({
    "name": ["Aarav", "Priya", "Charlie", "Bob", "Eve", "Adam"],
    "score": [90, 93, 69, 74, 77, 89],
    "subject": ["Math", "Science", "Math","Science", "Math", "Science"],
    "city": ["Kathmandu", "POkhara", "Kathmandu", "Biratnagar", "Pokhara", "Kathmandu"]
    })

print(f"The dataframe is listed as:\n{df}\n")

# groppby both subject and city
print(df.groupby(["subject", "city"])["score"].mean())

# count students per subject per city
print(df.groupby(["subject", "city"])["score"].count())

# Reset index to flatten result
print(df.groupby(["subject", "city"])["score"].mean().reset_index())

```

```python
# Advanced Filtering
# multiple conditions, isin, str.contains
# beyod basic boolean filters. three patterns come up constantly

# & means AND and | means OR (always wrap each condition in parentheses)
# isin(list) -> checks if value is in a list
# str.contains() -> filters text columns by substring
# ~(tilde) -> NOT, inverts any filter

# Always wrap each condition in () when combining with & or |.
# Forgetting this is the #1 filtering bugs in Pandas
```

```python
df = pd.DataFrame({
    "name": ["Aarav", "Bob", "Charlie", "Donald", "Eve", "Figo"],
    "score" : [89, 88, 50, 67, 77, 90],
    "subject" : ["Maths", "Sciecne", "Maths" ,"Science", "Maths", "Science"],
    "city": ["Kathmandu", "Pokhara", "Biratnagar", "Pokhara", "Kathmandu", "Pokhara"]

})

# AND OR operations
print(df[(df["score"]>75) & (df["subject"]== "Maths")])
print(df[(df["city"]== "Kathmandu") | (df["city"]=="Pokhara")])

# isin() -> cleaner than chained OR
print(df[df["city"].isin(["Kathmandu", "Pokhara"])])
print("\n")

# str.contains -> text speech
print(df[df["name"].str.contains("a", case = False)])
print("\n")

# ~ means NOT
print(df[~df["subject"].str.contains("Maths")])
```

```python
# sort values and ranking
# ordering and ranking your data
# sort_values() -> sorts a DAtaFrame by one or more columns.
# You can sort multiple columns with differnet directions -> very useful for leaderboard and reports

# rank() -> assigns a rank number to each row based on a column value/

# ascending = False gives you descending order, this is want you want for scores, sales, rankings
```

```python
df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie", "Duck", "Eva" ],
    "score" :[88, 89, 60, 64, 77],
    "subject": ["Maths", "Science", "English", "Maths", "Science"],
    "city": ["Kathmandu", "Pokhara", "Kathmandu", "Chitwan", "Kathmandu"]

})

# sort by score descending 
df_sorted = df.sort_values("score", ascending= False)
print(df_sorted,"\n")

# dort by multiple columns
df_sorted = df.sort_values(["subject", "score"], ascending=[False, True])
print(df_sorted,"\n")

# add rank column
df["rank"] = df["score"].rank(ascending = True).astype(int)
print(df.sort_values("rank"))
```

```python
# sorting file by name alphabeticlly, then by marks descending within each grade and adding a rank column based on marks

df = pd.read_csv("learn.csv")
print(df.isnull().sum())

df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
df["Passed"] = df["Passed"].fillna("Unknown")

print(f"DataFrame:\n{df}\n")

df_sorted = df.sort_values("Name", ascending = True)
print(f"The name is alphabetically order:\n{df_sorted}\n")

df_sorted = df.sort_values("Marks", ascending = False)
print(f"The Marks in descending order:\n{df_sorted}\n")

df["Rank"] = df["Marks"].rank(ascending = False).astype(int)
print(df.sort_values("Rank"))
```

```python
# pivot_table : reshaping data into a summary grid
# pivot_table reshapes your data into a grid -> rows are one category, columns are another, and the cells contain an aggregated values.
# this is exactly what Excel pivots table de, but in one line of code. Perfect for comparing two categorical variables.

# aggfun can be 'mean', 'sum', 'count', 'max', 'min' or even a list of functions/
# fill_value=0 fills any empty cells with 0 insted of NaN
```

```python
df = pd.DataFrame({
    "name": ["Aarav", "Bob", "Charlie", "Donald", "Eve", "Figo"],
    "score" : [89, 88, 50, 67, 77, 90],
    "subject" : ["Maths", "Science", "Maths" ,"Science", "Maths", "Science"],
    "city": ["Kathmandu", "Pokhara", "Biratnagar", "Pokhara", "Kathmandu", "Pokhara"]

})

# average score per city per subject
pivot = df.pivot_table(
    values = "score",
    index = "city", 
    columns = "subject",
    aggfunc = "mean"
)
print(pivot,"\n")

#count of students per city per subject
pivot2= df.pivot_table(
    values ="name",
    index = "city", 
    columns = "subject",
    aggfunc = "count",
    fill_value= 0
)
print(pivot2,"\n")

# max scoer per city per subject
pivot3= df.pivot_table(
    values = "score",
    index = "city",
    columns = "subject",
    aggfunc = "max"
)
print(pivot3,"\n")

# count od students per city
pivot4 = df.pivot_table(
    values = "name",
    index = "city",
    columns = "subject",
    aggfunc = "count",
    fill_value = 0
)
print(pivot4)
```

```python
# mini analysis -> putting it all together
# a real workflow combinig all of Pandas

# this is exact workflow -> load, clean, group, filter, summarize -> is what data analyst do every single day.
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "name": ["Aarav", "Priya", "Sita", "Ram", "Rijan", "Gita", "Hari", "Mina"],
    "subject": ["Maths", "Science", "Maths", "Science", "Maths", "Science", "Maths", "Science"],
    "score": [88, 72, 95, 61, 85, np.nan, 91, np.nan],
    "city": ["KTM", "PKR", "KTM", "BRT", "PKR", "KTM", "BRT", "PKR"]

})

# cleaning nulls
df["score"] = df["score"].fillna(df["score"].mean())

# add grade 
df["grade"] = df["score"].apply(
    lambda x: "A" if x>=85 else "B" if x>=70 else "C"
)

#group analysis
print("Average by subject:\n")
print(df.groupby("subject")["score"].mean().round(1),"\n")

# top city
city_avg = df.groupby("city")["score"].mean().round(1)
print(f"Top city:{city_avg.idxmax()}\n")

# filter top performers
top = df[df["score"]>85]
print(f"Top performers: {len(top)} students")
print(top[["name", "subject", "city", "score", "grade"]])
print("\n")

# pivot summary
pivot = df.pivot_table(
    values = "score", 
    index = "city",
    columns = "subject",
    aggfunc = "mean",
    fill_value = 0
)
print(f"Pivot table:\n{pivot}\n")

print(df.groupby(["subject","grade"])["name"].count(),"\n")

print(f"Lowest city: {city_avg.idxmin()}")
```
