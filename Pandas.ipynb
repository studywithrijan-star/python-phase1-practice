```python
# Selectind data
# accessing columns, rows and specific cells

# pandas has two main ways to select data:
# df["column"] -> selects a single column (return a Series)
# df[["col1", "col2"]] -> selects multiple columns (returns a DataFrane)
# df.loc[rowmcol] -> selects by label (bane/index label)
# df.iloc[rowm cil] -> selects by position (number)

# Boolean filtering works just like Numpy - pass a condition inside df[] to filter rows
# loc uses labels, iloc uses integers. When in doubt, loc is almost always what you wnat for named columns and indexes.
# ------------RIJAN----------------
```

```python
df = pd.DataFrame({
    "name": ["King", "Queen", "Jack", "Joker"],
    "score": [45, 56, 67, 78],
    "passed": [False, True, True, True],
    "grade": ["F", "D", "C", "B"],

})

print(f"DataFrame: \n {df} \n")

# single column -> series
print(df["name"])
print("\n")

# multiple column -> dataframe
print(df[["name", "grade"]])
print("\n")

#  filter rows -> student who passed
print(df[df["passed"] == True])
print("\n")

# filter score above 55
print(df[df["score"]>55])
print("\n")

# select by position
print(df.iloc[0])       # first row
print("\n")

print(df.iloc[0,1])   
print("\n")
# print(df.iloc[0:1, 0:2])   # first 2 rows and first 2 columns
```

```python
# Adding and modifying columns
# transforming your data
# you can add new columns to a DataFrame by assigning to a new column name. Pandas applies the operation ot every row automatically, vectorized just like Numpy.

# apply() lets you apply any custom function to each row or column - its the most flexible transformation tool in Pandas.
# You can also rename columns and drop one you dint need/

# df["new_col"] = ... is the stnadard way to ad column. It modifies the DataFrame in place
```

```python
df = pd.DataFrame({
    "name": ["Ace", "Spade", "Heart", "Diamond", "Joker"],
    "score": [67, 67, 89, 90, 99],
    "subject": ["Math", "Chemistry", "English", "Math", "English"],
    "city": ["Delhi", "Kathmandu", "Bhaktapur", "Lalitput", "Pokhara"]
})

print(f"DataFrame is as \n {df} \n")
# Add a new column
df["bonus"] = df["score"]*0.1
print(f"Dataframe with new column is: \n{df}\n")

# add column based on condition
df["status"] = df["score"].apply(
    lambda x: "pass" if x>=70 else "fail")
print(f"DataFrame with status column: \n{df}\n")

# rename a column
df = df.rename(columns={"score": "marks"})
print(f"Dataframe after renaming is: \n{df}\n")

# drop a column
df = df.drop(columns= ["status"])
print(f"Dataframe after dropping is: \n{df}\n")


```

```python
# creating a new dataframe

df = pd.DataFrame({
    "name": ["Jack", "Rohit", "Sandeep", "Nandan", "Sher"],
    "score": [77, 78, 89, 5, 91],
    "city": ["Mumbai", "Lumbini", "Biratnagar", "Karnali", "Lumbini"],
    
})

print(f"The Dataframe is: \n{df}\n")

# adding new column grade
df["grade"] = df["score"].apply(
    lambda x : "A" if x >=85 else "B" if x>=70 else "C"
)
print(f"Dataframe with grade is: \n{df}\n")

## calulating percentage
df["percentage"] = (df["score"] /120*100).round(1)
print(f"The dataframe with percentage is: \n{df}\n")
```

```python
# Handling missing data
# the most common real-world problem

# real datasets alwayss have missinf values represented as Nan(Not a Number) in Pandas. Handling them correctly is one of the most important data cleaning skills.

# isnull() -> detects missing values
# dropna() -> removes rows with any Nan
# fillna(value) -> fills NaN with a value
# fillna(df.mean()) -> replace with column average

# always check for nulls early, they silently break calculations

# df.isnull().sum() is the fastest way to see how many missing values exists in each column.
#  Run it on every new dataset
```

```python
# creating a dataframe with missing values

df = pd.DataFrame({
    "name": ["Aarav", "Priya", "Sita", "Ram"],
    "score": [88, np.nan, 96, 61],
    "city": ["KTM","PKR",np.nan,"BRT"]
})

# Count missing values per column
print(df.isnull().sum())

# Fill numeric NaN with column mean
df["score"] = df["score"].fillna(df["score"].mean().round(1)) 

# Fill string NaN with a placeholder
df["city"] = df["city"].fillna("Unknown")

print(f"The DataFrame after handling missing values: \n{df}\n")
```

```python
# creating one more new dataframe with missing values and handling them

df = pd.DataFrame({
    "name": ["Aarav", "Adam", np.nan, "Charlie"],
    "score": [88, 75, np.nan, np.nan],
    "city" : ["Bhaktapur", np.nan, np.nan, "Kritipur"]

})

print(f"Dataframe with missing values: \n{df}\n")

print(df.isnull().sum())
print("\n")

df["name"] = df["name"].fillna("Unknown")
df["score"] = df["score"].fillna(df["score"].mean())
df["city"] = df["city"].fillna("Unknown")
print(f"The DataFrame after handling missing data:\n{df}\n")
```
